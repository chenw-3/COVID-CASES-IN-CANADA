# Data Wrangling Template

## Gather

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading file and getting a first look of structure
df = pd.read_csv('cases.csv', low_memory=True)
df.head(50)
```

## Assess

```python
#looking at the no. of rows
df.shape
```

## Clean

```python
#extracting important columns
df1 = df[['province', 'date_report', 'travel_yn', 'locally_acquired', 'additional_info']]
df1
```

```python
#removing all rows with 'Not reported'
df1 = df1[df1['travel_yn'] != 'Not Reported']
df1
```

```python
#changing travel variable to numeric
df1["travel_yn"] = df1["travel_yn"].apply(pd.to_numeric)
```

```python
#looking at the values in the locally_acquired column
df1.locally_acquired.value_counts()
```

```python
#making community spread column
community_spread = df1.locally_acquired.str.extract('(Community)', expand=True)
community_spread
```

```python
#adding community spread column to df1
df1['community_spread'] = community_spread
df1
```

```python
#making all close contact variables the same
close = df1.locally_acquired.str.extract('([Cc]lose\s[Cc])')
close.value_counts()
```

```python
#adding close contact coumn to df1
df1['close_contact'] = close.replace({'Close C': 'close', 'Close c': 'close', 'close c': 'close'})
df1
```

```python
#dropping locally acquired column
df1 = df1.drop(['locally_acquired', 'additional_info'], axis = 1)
df1
```

```python
#checking values
df1.info()
```

```python
#getting variables as 1's and 0's
df1 = pd.get_dummies(df1, columns=['travel_yn','community_spread', 'close_contact'])
df1
```

```python
#renaming columns
df1 = df1.rename(columns={'community_spread_Community': 'community_spread', 'close_contact_close': 'close_contact'})
```

```python
df1 = df1.drop('travel_yn_0', axis=1)
df1
```

```python
df1.mean()
```

```python
#grouping by province
province = df1.groupby(['date_report'], as_index=True)['province'].value_counts()
province
```

```python
#writing to a csv
df1.to_csv('COVID_cases.csv')
```
